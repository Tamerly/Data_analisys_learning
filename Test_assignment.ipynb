После изучения структуры данных начинаю анализ.

> Нахожу топ 10 товаров с наибольшей выручкой:



```python
product_total_revenue = df.groupby('producttitle', as_index=False) \
    .agg({'revenue': 'sum'}) \
    .sort_values('revenue', ascending=False)

product_total_revenue.head(10)
```

Проверяю количество проданных единиц на каждый из продуктов:

```python
df.groupby('producttitle', as_index=False) \
    .agg({'revenue': 'count'}) \
    .sort_values('revenue', ascending=False)
```

Смотрю, какой продукт старые пользователи покупают чаще:

```python
df.query('user_type == "old"') \
    .groupby('producttitle', as_index=False) \
    .agg({'revenue': 'count'}) \
    .sort_values('revenue', ascending=False)
```

Выясняю - какие пользователи приноят больше выручки через Телеграм:

```python
df.query('Source == "Телеграм"') \
    .groupby('user_type') \
    .agg({'revenue': 'sum'}) \
    .sort_values('revenue', ascending=False)
```
